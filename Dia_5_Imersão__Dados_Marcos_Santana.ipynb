```python
# Distribuição de Frequência para a variável "tratamento".
frequencia = dados['tratamento'].value_counts()
percentual = dados['tratamento'].value_counts(normalize = True) * 100

# Crindo uma tabela de frequência para tratamento.
dist_freq_tratamento = pd.DataFrame({
    "Frequencia" : frequencia,
    "Porcentagem (%)" : percentual
} )
dist_freq_tratamento.head()
```

```python
# Distribuição de Frequência para a variável "tempo".
frequencia = dados['tempo'].value_counts()
percentual = dados['tempo'].value_counts(normalize = True) * 100

# Criando uma tabela Frequência e Porcentagem "tempo".
dist_freq_tempo = pd.DataFrame({
    "Frequencia" : frequencia,
    "Porcentagem (%)" : percentual
})
dist_freq_tempo.sort_index()
```

```python
# Distribuição de Frequência para a variável "dose".
frequencia = dados['dose'].value_counts()
percentual = dados['dose'].value_counts(normalize = True) * 100

# Criando uma tabela Frequência e Porcentagem "dose".
dist_freq_dose = pd.DataFrame({
    "Frequencia" : frequencia,
    "Porcentagem (%)" : percentual
})
dist_freq_dose.head()
```

```python
# Distribuição de Frequência para a variável "droga".
frequencia = dados['droga'].value_counts()
percentual = dados['droga'].value_counts(normalize = True) * 100

# Criando uma tabela Frequência e Porcentagem "dose".
dist_freq_droga = pd.DataFrame({
    "Frequencia" : frequencia,
    "Porcentagem (%)" : percentual
})
dist_freq_droga.head()
```

```python
# Criando uma correlação entre duas variáveis do dataset.
pd.crosstab(dados['dose'], dados['tempo'])
```

Então analisando nossa tabela de frequência, temos os dados, onde dose e tempo tem uma correlação, mas não sabemos que tipo de tratamento foi utilizado.

Agora conseguimos analisa a tabela através do tipo de tratamento.

```python
# Normalizando tabela frequência
pd.crosstab([dados['dose'], dados['tempo']], dados['tratamento'], normalize = 'index')
```

Fazer este tipo de análise possibilita que façamos algumas suposições acerca do balanceamento entre as categorias e, analisando a nossa matriz, podemos concluir que há proporcionalidade na nossa base de dados.

# Dia 5 - Machine Learning, Sci-kit Learning e Desafios envolvidos

## Modelo de Regreção Lógica (Logistic Regression)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size= 0.2,
                                                        random_state= 373, stratify= y)

modelo_rlogistica = LogisticRegression(solver='newton-cg')
modelo_rlogistica.fit(x_treino , y_treino)
modelo_rlogistica.score(x_teste, y_teste)
```

## Modelo Dummy

### Modele de treino e teste em repetição dentro de um range
